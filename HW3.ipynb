```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

iris = load_iris()
X = iris.data
y = iris.target

k_graph_1 = range(1, 31)
k_scores1 = []

for k in k_graph_1:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10)
    k_scores1.append(scores.mean())

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_graph_1, k_scores1)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross Validated Accuracy')


k_graph_2 = range(11, 16)
k_scores2 = []


for k in k_graph_2:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10)
    k_scores2.append(scores.mean())

plt.subplot(1, 2, 2)
plt.plot(k_graph_2, k_scores2)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross Validated Accuracy')
plt.tight_layout()
plt.show()

param_grid = {'n_neighbors': k_graph_2}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=10, scoring='accuracy')
grid.fit(X, y)
print("The optimal k value is:", grid.best_params_['n_neighbors'])
```
