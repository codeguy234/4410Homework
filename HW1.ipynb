```python
url = 'https://vincentarelbundock.github.io/Rdatasets/csv/ggplot2/diamonds.csv'

import pandas as pd
import io
import matplotlib.pyplot as plt


df = pd.read_csv(url)
print(df.head (7))
print(df.tail (7))

stats = df[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].describe()

print(stats)

cut = df['cut'].describe()
color = df['color'].describe()
clarity = df['clarity'].describe()

print("Cut stats:\n", cut, "\n")
print("Color stats:\n", color, "\n")
print("Clarity stats:\n", clarity)

cut_uni = df['cut'].unique()
color_uni = df['color'].unique()
clarity_uni = df['clarity'].unique()

print("Cut unique values:", cut_uni)
print("Color Unique values:", color_uni)
print("Clarity unique values:", clarity_uni)

df[['carat', 'depth', 'price', 'table', 'x', 'y', 'z']].hist(figsize=(12, 8), bins=30, edgecolor='black')
plt.suptitle("Histograms", fontsize=20)
plt.show()
```
